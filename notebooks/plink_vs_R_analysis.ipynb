```python
def plot_agreement(df, phe):
    plt.scatter(df["LOG OR"], df["LOG OR (R)"])
    plt.xlabel("log OR (PLINK)")
    plt.ylabel("log OR (R)")
    plt.title("plot of agreement between R and PLINK odds ratios for " + phe)
    plt.savefig("plots/" + phe + "_R_PLINK_plot.png")
    #plt.show()
    plt.close()
```

## Create concordance plots for all phenotypes

```python

wrong_phes = []
for i in range(446):
    phe = "HC" + str(i)
    try:
        df = pd.read_csv(path + "scripts/output/process_results/" + phe + "_processed" + 
                         firth_toggle + ".txt", header=0, delimiter='\t')
        plot_agreement(df, phe)
    except:
        wrong_phes.append(phe)
        
print(wrong_phes)
```

## Examine points of discordance

```python


num = 446
hap_discord = {}
phe_discord = [0 for i in range(num)]
diff_val = []
thresh = 0.1
for i in range(num):
    if i != 29 and i != 65:
        phe = "HC" + str(i)
        #try:
        df = pd.read_csv(path + "scripts/output/process_output/" + phe + "_processed" + firth_toggle 
                         + ".txt", header=0, delimiter='\t')
        #print(df["DIFF"])
        haps = df.loc[df["OR DIFF"] > thresh]["ID"].tolist()
        diff_val = diff_val + df.loc[df["OR DIFF"] > thresh]["OR DIFF"].tolist()
        phe_discord[i] = len(haps)
        for h in haps:
            if h not in hap_discord.keys():
                hap_discord[h] = 1
            else:
                hap_discord[h] += 1
        
    #except:
     #   wrong_phes.append(phe)
        
#print(phe_discord)
#print(len(hap_discord.keys()))
#print(hap_discord)
diff_val_2 = [x for x in diff_val if x < 2]
#print(diff_val)
plt.hist(diff_val_2)
#plt.xlim([0,20000])
plt.title('difference in the odds ratios for values between 0.1 and 2 (' + str(len(diff_val_2)) 
          + ' out of ' + str(len(diff_val)) + ')')
plt.xlabel('difference of OD')
plt.ylabel('number of occurences')
plt.show()
print(len(diff_val_2)/float(len(diff_val)))
```

```python
plt.hist(phe_discord,8)
plt.title('number of discordant haplotypes for each phenotype')
plt.xlabel('number discordant haplotypes')
plt.ylabel('number of phenotypes')
plt.savefig(outpath + 'phe_discord' + firth_toggle + '.png')
plt.show()
plt.close()
#print(hap_discord.items())
hap = []
val = []
print(len(hap_discord.keys()))
#hap_val = np.array(shape=(len(hap_discord.keys()),2))

#print(hap_val)
for key in hap_discord.keys():
    hap.append(key)
    val.append(hap_discord[key])
    #hap_val[i][0] = key
    #hap_val[i][1] = hap_discord[key]
    
df = pd.DataFrame({"haplotype" : hap, "number of discordances" : val})
df = df.sort_values(["number of discordances"])
df = df.set_index("haplotype")
#print(df)
#print(df["val"])
#print(df["hap"])
df.plot.bar()
plt.title("frequency of there being a discordance > " + str(thresh) + " between OR for PLINK and R")
plt.ylabel("number of times discordance > " + str(thresh))
#plt.bar(hap,val)
#plt.bar(df["hap"],df["val"])
plt.xticks(rotation=90)
fig = plt.gcf()
fig.set_size_inches(18.5, 10.5)
plt.savefig(outpath + 'freq_discord' + firth_toggle + '.png')
plt.show()
plt.close()
```

## PLINK OD plot

```python
full_df = full_df.sort_values('OR')
ind_hap = []
plt.scatter(range(full_df.shape[0]),full_df['LOG OR'],c='royalblue')
highlight = True

if highlight:
    for i in range(len(phe_hap)):
        hap_ind = list(full_df['ID']).index(phe_hap[i])
        ind_hap.append(hap_ind)
        plt.scatter(hap_ind,list(full_df['LOG OR'])[hap_ind], c='gold')
plt.xlabel('sorted haplotype')
plt.ylabel('log OR (PLINK)')
plt.title('log odds ratio (PLINK) with key haplotypes highlighted for ' + phe_name)
ax = plt.gca()
ax.set_xticks(ind_hap)
ax.set_xticklabels(phe_hap)
plt.xticks(rotation=90,fontsize=10)
fig = plt.gcf()
fig.set_size_inches(10, 8)
plt.savefig(outpath + curr_phe + '_OR_PLINK_plot' + firth_toggle + '.png')
plt.show()

```
