# Get interaction p values

```python
from collections import defaultdict

phe_allelotype_dict = defaultdict(list)

for file in os.listdir("../scripts/output/interact/"):
    if file.endswith(".txt"):
        phe = file.split("_")[0]
        f = open("../scripts/output/interact/" + file)
        print(file)
        for line in f.readlines():
            split_line = line.split()
            if line[0] == "$":
                inter = line[2:-2]
                phe_allelotype_dict[phe].append(inter)
```

```python
#lst=[zip([k] * len(v), v) for k, v in tmp.items()]

phe_list = [[k] * len(v) for k, v in tmp.items()]
allelotype_list = [v for k, v in tmp.items()]

phe_list = [item for sublist in phe_list for item in sublist]
allelotype_list = [item for sublist in allelotype_list for item in sublist]
```

```python
big_fat_df = pd.DataFrame(index = [np.array(phe_list), np.array(allelotype_list)], columns = ['estimate_add1', 'se_add1', 'z_add1', 'p_add1', 'estimate_add2', 'se_add2', 'z_add2', 'p_add2', 'estimate_int', 'se_int', 'z_int', 'p_int'])
```

```python
big_fat_df
```

```python
t0 = time.time()
count = 1

def get_pval(split_line, e_and_p, inter):
    if e_and_p:
        p_val = split_line[4]
    else:
        p_val = split_line[1]
    if p_val == "<2e-16" or p_val == "<":
        p_val = 2e-16
    else:
        p_val = float(p_val)
    return p_val

def assign_estimate_se_z(split_line, big_fat_df, typestring, e_and_p, inter):
    #print(typestring)
    if e_and_p:
        big_fat_df.loc[(phe, inter)]['p_' + typestring] = get_pval(split_line, e_and_p, inter)
        #print('p = ' + split_line[4])
    big_fat_df.loc[(phe, inter)]['estimate_' + typestring] = split_line[1]
    #print('estimate = ' + split_line[1])
    #print(big_fat_df.loc[(phe, inter)]['estimate_' + typestring])
    big_fat_df.loc[(phe, inter)]['se_' + typestring] = split_line[2]
    #print('se = ' + split_line[2])
    big_fat_df.loc[(phe, inter)]['z_' + typestring] = split_line[3]
    #print('z = ' + split_line[3])
    return big_fat_df

for file in os.listdir("../scripts/output/interact/"):
    if file.endswith(".txt"):
        phe = file.split("_")[0]
        f = open("../scripts/output/interact/" + file)
        print(count)
        count += 1
        print(file)
        print("time: {}\n".format(time.time() - t0))
        for line in f.readlines():
            split_line = line.split()
            if line[0] == "$":
                inter = line[2:-2]
                #print(inter)
                p_line = False
                e_and_p = False
                
            if len(split_line) > 0:

                if split_line[0] == "Estimate":
                    if "Pr(>|z|)" in line:
                        e_and_p = True
                    else:
                        e_and_p = False
                    #print('enp: '+ str(e_and_p))

                elif split_line[0] == "Pr(>|z|)":
                    p_line = True

                if e_and_p:
                    if split_line[0] == "as.numeric(gcounts1)" and len(split_line) > 4:
                        big_fat_df.loc[(phe, inter)]['p_add1'] = get_pval(split_line, e_and_p, inter)
                        big_fat_df = assign_estimate_se_z(split_line, big_fat_df, 'add1', e_and_p, inter)
                    elif split_line[0] == "as.numeric(gcounts2)" and len(split_line) > 4:
                        big_fat_df.loc[(phe, inter)]['p_add2'] = get_pval(split_line, e_and_p, inter)
                        big_fat_df = assign_estimate_se_z(split_line, big_fat_df, 'add2', e_and_p, inter)
                    elif split_line[0] == "as.numeric(gcounts1):as.numeric(gcounts2)" and len(split_line) > 4:
                        big_fat_df.loc[(phe, inter)]['p_int'] = get_pval(split_line, e_and_p, inter)
                        big_fat_df = assign_estimate_se_z(split_line, big_fat_df, 'int', e_and_p, inter)

                else:
                    if split_line[0] == "as.numeric(gcounts1)":
                        if p_line:
                            #print(split_line)
                            big_fat_df.loc[(phe, inter)]['p_add1'] = get_pval(split_line, e_and_p, inter)
                            #print('p = ' + str(get_pval(split_line, e_and_p, inter)))
                        else:
                            big_fat_df = assign_estimate_se_z(split_line, big_fat_df, 'add1', e_and_p, inter)

                    elif split_line[0] == "as.numeric(gcounts2)":
                        if p_line:
                            big_fat_df.loc[(phe, inter)]['p_add2'] = get_pval(split_line, e_and_p, inter)
                            #print('p = ' + str(get_pval(split_line, e_and_p, inter)))
                        else:
                            big_fat_df = assign_estimate_se_z(split_line, big_fat_df, 'add2', e_and_p, inter)

                    elif split_line[0] == "as.numeric(gcounts1):as.numeric(gcounts2)":
                        if p_line:
                            big_fat_df.loc[(phe, inter)]['p_int'] = get_pval(split_line, e_and_p, inter)
                            #print('p = ' + str(get_pval(split_line, e_and_p, inter)))
                        else:
                            big_fat_df = assign_estimate_se_z(split_line, big_fat_df, 'int', e_and_p, inter)            
                            p_line = False
        f.close()
print("time: {}".format(time.time() - t0))
```

```python
from statsmodels.stats import multitest

big_fat_df['adj_p_add1'] = multitest.multipletests(big_fat_df['p_add1'], is_sorted=False, method='fdr_by')
big_fat_df['adj_p_add1'] = multitest.multipletests(big_fat_df['p_add1'], is_sorted=False, method='fdr_by')
big_fat_df['adj_p_add1'] = multitest.multipletests(big_fat_df['p_add1'], is_sorted=False, method='fdr_by')
```

```python
big_fat_df.to_csv(outpath + "interaction_results.csv")
```

# Adjust pvalues

```python
thresh = 0.05
s = big_fat_df[['p_add1', 'p_add2', 'p_int']].as_matrix().shape
flat_df = big_fat_df[['p_add1', 'p_add2', 'p_int']].as_matrix().flatten()
total = flat_df.shape[0]
nom_p = robjects.FloatVector(flat_df)
rpadjust = robjects.r['p.adjust']
adj_p = np.array(rpadjust(nom_p, "BY"))
adj_p = np.reshape(adj_p,s)
# find the number of adjusted values below the threshold
below_thresh = (adj_p < thresh).sum()
adj_p_df = pd.DataFrame(adj_p, index=big_fat_df[['p_add1', 'p_add2', 'p_int']].index.values, columns=big_fat_df[['p_add1', 'p_add2', 'p_int']].columns.values)
print("number below threshold: {}".format(below_thresh))
adj_p_df.to_csv(outpath + "interact_pvals_adj.csv")
adj_p_df;
```

```python
sig_df = adj_p_df.loc[(adj_p_df < thresh).sum(axis=1) > 0, :]
sig_df = sig_df[sig_df.columns[(sig_df< thresh).any()]]
print("number of phenotypes with significant entry: {}".format(sig_df.shape[0]))
print("number of allelotypes with significant entry: {}".format(sig_df.shape[1]))
sig_df.to_csv(outpath + "interact_pvals_adj_sig.csv")
sig_df;
```

```python
sig_df[sig_df < thresh].count().sort_values(ascending = False).to_csv(outpath + "allele_sig_counts.csv")
T_sig_df = sig_df.transpose()
T_sig_df[T_sig_df < thresh].count().sort_values(ascending = False).to_csv(outpath + "phe_sig_counts.csv")
```
