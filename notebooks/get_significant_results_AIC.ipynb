```python
import pandas as pd
```

```python
#Drop the columns that are NaN, index the first column
gen_adj_p_1 = pd.read_csv('../manuscript/genotype_assoc_adj_p_all_1.csv', index_col = 0).dropna(axis='columns', how='all')
gen_adj_p_2 = pd.read_csv('../manuscript/genotype_assoc_adj_p_all_2.csv', index_col = 0).dropna(axis='columns', how='all')
#Looks like the AIC diff table has one extra haplotype not included in other tables, drop
aic_diff = pd.read_csv('../manuscript/AIC_diff.csv', index_col = 0).dropna(axis='columns', how='all').drop(columns=['B_1510'])
#get only the rows that are included in the other tables for AIC_diff
aic_diff = aic_diff[aic_diff.index.isin(gen_adj_p_1.index)]
```

```python
sig_aic_diff_1 = aic_diff[(aic_diff <= -1) & (gen_adj_p_1 < 0.05)].dropna(axis='columns', how='all').dropna(axis='rows', how='all')
sig_aic_diff_2 = aic_diff[(aic_diff <= -1) & (gen_adj_p_2 < 0.05)].dropna(axis='columns', how='all').dropna(axis='rows', how='all')
```

```python
sig_adj_p_1 = gen_adj_p_1[(aic_diff <= -1) & (gen_adj_p_1 < 0.05)].dropna(axis='columns', how='all').dropna(axis='rows', how='all')
sig_adj_p_2 = gen_adj_p_2[(aic_diff <= -1) & (gen_adj_p_2 < 0.05)].dropna(axis='columns', how='all').dropna(axis='rows', how='all')
```
